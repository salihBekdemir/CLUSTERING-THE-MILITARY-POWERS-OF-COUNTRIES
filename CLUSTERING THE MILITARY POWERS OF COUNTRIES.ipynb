```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel('World military power.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
columns_to_delete = df.columns[2::2]
df = df.drop(columns=columns_to_delete)
```

```python
df.head()
```

```python
new_columns = df.iloc[0]
df = df[1:]
```

```python
df.head()
```

```python
df.columns = new_columns
```

```python
df.head()
```

```python
df.info()
```

```python
object_columns = df.select_dtypes(include='object').columns.tolist()
object_columns.remove('Military Strength')
```

```python
for column in object_columns:
    df[column] = df[column].apply(lambda x: float(''.join(x.split(','))) if isinstance(x, str) else x)
```

```python
df[object_columns] = df[object_columns].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Navy Fleet Strengths value']=df['Navy Fleet Strengths value'].fillna(df['Navy Fleet Strengths value'].median())
df['Aircraft Carrier Fleet Strength value']=df['Aircraft Carrier Fleet Strength value'].fillna(df['Aircraft Carrier Fleet Strength value'].median())
df['Submarine Fleet Strength value']=df['Submarine Fleet Strength value'].fillna(df['Submarine Fleet Strength value'].median())
df['Destroyer Fleet Strength value']=df['Destroyer Fleet Strength value'].fillna(df['Destroyer Fleet Strength value'].median())
df['Frigate Fleet Strength value']=df['Frigate Fleet Strength value'].fillna(df['Frigate Fleet Strength value'].median())
df['Total Coastline Coverage value']=df['Total Coastline Coverage value'].fillna(df['Total Coastline Coverage value'].median())
df['Total Border Coverage value']=df['Total Border Coverage value'].fillna(df['Total Border Coverage value'].median())
```

```python
df.isnull().sum()
```

```python
x = df.drop("Military Strength", axis=1)
```

```python
x
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
ss=[]
for i in range(2,10):
    model=KMeans(i)
    model=model.fit(x)
    tahmin=model.predict(x)
    ss1=silhouette_score(x,tahmin)
    ss.append(ss1)
    print(ss1)
    wcss.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(2,10),wcss)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=range(2,10),y=wcss,marker='*')
```
